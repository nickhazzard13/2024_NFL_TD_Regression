```python
import pandas as pd
```

```python
df = pd.read_csv('../files/2024_fantasy_stats.csv')
df.head()
```

Renaming Columns

```python
df.rename(columns= {'YDS':'Pass_YDS', 'TD': 'Pass_TD', 'ATT' : 'Rush_ATT','YDS.1': 'Rush_YDS', 'TD.1':'Rush_TD', 'YDS.2':'Rec_YDS', 'TD.2':'Rec_TD'}, inplace = True)
```

Additional Stats for Analysis -- Must have at least 9 GP

```python
df['INT_pergame'] = (df['INT'] / df['GP']).round(2)
df['YDS_percomp'] = (df['Pass_YDS'] / df['GP'] ).round(2)
df['YDS_percarry'] = (df['Rush_YDS'] / df['Rush_ATT']).round(2)
df['Rush_TD_pergame'] = (df['Rush_TD'] / df['GP']).round(2)
df['YDS_perrec'] = (df['Rec_YDS'] / df['REC']).round(2)
df['TGT_pergame'] = (df['TGT'] / df['GP']).round(2)
df['Rec_TD_pergame'] = (df['Rec_TD'] / df['GP']).round(2)
df['Catch_rate'] = (df['REC'] / df['TGT'] ).round(2)
df['PTS_fromTD'] = ((df['Pass_TD']*4 + df['Rush_TD']*6 + df['Rec_TD']*6)/ (df['PTS'])).round(2)
df = df[df['GP'] >= 9].reset_index(drop=True)
df.replace([float('inf'), float('-inf')], 0, inplace=True)
df.fillna(0, inplace = True)

df.head()
```

Positional DF Creation


```python
df_qb = df[df['POS'] == 'QB'].reset_index(drop=True)
df_rb = df[df['POS'] == 'RB'].reset_index(drop=True)
df_wr = df[df['POS'] == 'WR'].reset_index(drop=True)
df_te = df[df['POS'] == 'TE'].reset_index(drop=True)

df_wr.head()
```

VOL-Based Expected Fantasy (League Average)

RBs

```python
YPC_rb = round((df_rb['Rush_YDS'].sum() / df_rb['Rush_ATT'].sum()),2)
TD_perrush_rb = round((df_rb['Rush_TD'].sum() / df_rb['Rush_ATT'].sum()),2)
rb_catch_rate = round((df_rb['REC'].sum() / df_rb['TGT'].sum()),2)
rb_ypt = round((df_rb['Rec_YDS'].sum() / df_rb['TGT'].sum()),2) 
rb_td_per_tgt = round((df_rb['Rec_TD'].sum() / df_rb['TGT'].sum()),2)  

df_rb['exp_rush_yds'] = (df_rb['Rush_ATT'] * YPC_rb).round(2)
df_rb['exp_rush_td'] = (df_rb['Rush_ATT'] * TD_perrush_rb).round(2)
df_rb['exp_rec'] = (df_rb['TGT'] * rb_catch_rate).round(2)
df_rb['exp_rec_yds'] = (df_rb['TGT'] * rb_ypt).round(2)
df_rb['exp_rec_td'] = (df_rb['TGT'] * rb_td_per_tgt).round(2)
df_rb['exp_fantasy'] = (df_rb['exp_rush_yds']*0.1 + df_rb['exp_rush_td']*6 + df_rb['exp_rec']*1 + df_rb['exp_rec_yds']*0.1 + df_rb['exp_rec_td']*6).round(2)
df_rb['pt_diff'] = (df_rb['PTS'] - df_rb['exp_fantasy']).round(2)
```

WRs

```python
wr_catch_rate = round((df_wr['REC'].sum() / df_wr['TGT'].sum()),2)
wr_ypt = round((df_wr['Rec_YDS'].sum() / df_wr['TGT'].sum()),2) 
wr_td_per_tgt = round((df_wr['Rec_TD'].sum() / df_wr['TGT'].sum()),2)
wr_ypc = round((df_wr['Rush_YDS'].sum() / df_wr['Rush_ATT'].sum()),2)
wr_td_perrush = round((df_wr['Rush_TD'].sum() / df_wr['Rush_ATT'].sum()),2)

df_wr['exp_rec'] = (df_wr['TGT'] * wr_catch_rate).round(2)
df_wr['exp_rec_yds'] = (df_wr['TGT'] * wr_ypt).round(2)
df_wr['exp_rec_td'] = (df_wr['TGT'] * wr_td_per_tgt).round(2)
df_wr['exp_rush_yds'] = (df_wr['Rush_ATT'] * wr_ypc).round(2)
df_wr['exp_rush_td'] = (df_wr['Rush_ATT'] * wr_td_perrush).round(2)
df_wr['exp_fantasy'] = (df_wr['exp_rush_yds']*0.1 + df_wr['exp_rush_td']*6 + df_wr['exp_rec']*1 + df_wr['exp_rec_yds']*0.1 + df_wr['exp_rec_td']*6).round(2)
df_wr['pt_diff'] = (df_wr['PTS'] - df_wr['exp_fantasy']).round(2)   
```

TEs

```python
te_catch_rate = round((df_te['REC'].sum() / df_te['TGT'].sum()),2)
te_ypt = round((df_te['Rec_YDS'].sum() / df_te['TGT'].sum()),2) 
te_td_per_tgt = round((df_te['Rec_TD'].sum() / df_te['TGT'].sum()),2)    
te_ypc = round((df_te['Rush_YDS'].sum() / df_te['Rush_ATT'].sum()),2)
te_yds_per_rush = round((df_te['Rush_TD'].sum() / df_te['Rush_ATT'].sum()),2)   

df_te['exp_rec'] = (df_te['TGT'] * te_catch_rate).round(2)
df_te['exp_rec_yds'] = (df_te['TGT'] * te_ypt).round(2)
df_te['exp_rec_td'] = (df_te['TGT'] * te_td_per_tgt).round(2)
df_te['exp_rush_yds'] = (df_te['Rush_ATT'] * te_ypc).round(2)
df_te['exp_rush_td'] = (df_te['Rush_ATT'] * te_yds_per_rush).round(2)
df_te['exp_fantasy'] = (df_te['exp_rush_yds']*0.1 + df_te['exp_rush_td']*6 + df_te['exp_rec']*1 + df_te['exp_rec_yds']*0.1 + df_te['exp_rec_td']*6).round(2)
df_te['pt_diff'] = (df_te['PTS'] - df_te['exp_fantasy']).round(2)   
```

QBs

```python
yds_per_comp = round((df_qb['Pass_YDS'].sum() / df_qb['CMP'].sum()),2)
td_per_comp = round((df_qb['Pass_TD'].sum() / df_qb['CMP'].sum()),2)
int_per_comp = round((df_qb['INT'].sum() / df_qb['CMP'].sum()),2)
qb_ypc = round((df_qb['Rush_YDS'].sum() / df_qb['Rush_ATT'].sum()),2)
qb_td_perrush = round((df_qb['Rush_TD'].sum() / df_qb['Rush_ATT'].sum()),2)

df_qb['exp_pass_yds'] = (df_qb['CMP'] * yds_per_comp).round(2)
df_qb['exp_pass_td'] = (df_qb['CMP'] * td_per_comp).round(2)
df_qb['exp_ints'] = (df_qb['CMP'] * int_per_comp).round(2)
df_qb['exp_rush_yds'] = (df_qb['Rush_ATT'] * qb_ypc).round(2)
df_qb['exp_rush_td'] = (df_qb['Rush_ATT'] * qb_td_perrush).round(2)
df_qb['exp_fantasy'] = (df_qb['exp_pass_yds']*0.04 + df_qb['exp_pass_td']*4 - df_qb['exp_ints']*2 + df_qb['exp_rush_yds']*0.1 + df_qb['exp_rush_td']*6).round(2)
df_qb['pt_diff'] = (df_qb['PTS'] - df_qb['exp_fantasy']).round(2)
```

Additional Metrics

```python
df_rb['FP_pertouch'] = (df_rb['PTS'] / (df_rb['Rush_ATT'] + df_rb['REC'])).round(2)
df_rb['FP_percarry'] = (df_rb['PTS'] / df_rb['Rush_ATT']).round(2)
df_wr['FP_pertouch'] = (df_wr['PTS'] / (df_wr['Rush_ATT'] + df_wr['REC'])).round(2)
df_wr['FP_pertgt'] = (df_wr['PTS'] / df_wr['TGT']).round(2)
df_te['FP_pertouch'] = (df_te['PTS'] / (df_te['Rush_ATT'] + df_te['REC'])).round(2)
df_te['FP_pertgt'] = (df_te['PTS'] / df_te['TGT']).round(2)
```

```python
df_rb.to_csv('../files/2024_fantasy_rb_stats.csv', index = False)
df_wr.to_csv('../files/2024_fantasy_wr_stats.csv', index = False)
df_te.to_csv('../files/2024_fantasy_te_stats.csv', index = False)
df_qb.to_csv('../files/2024_fantasy_qb_stats.csv', index = False)

df.to_csv('../files/2024_fantasy_stats_all_postions.csv', index = False)
```
